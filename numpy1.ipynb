```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
print(a * b)
print(np.sin(a))

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(np.dot(A, B))
print(A @ B)
```

```python
arr = np.array([1, 2, 3, 4])
print(arr.sum())

print(arr.mean())
print(arr.max())
print(arr.min())
print(arr.std())
```
