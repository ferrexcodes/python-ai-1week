```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, color='purple', linestyle='-.', marker='s', linewidth=2)
plt.title('Customized Plot')
plt.xlabel('Index')
plt.ylabel('Values')
plt.grid(True)   # Adds grid lines
plt.show()
```

```python
x = [1, 2, 3]          

y1 = [1, 4, 9]         
y2 = [1, 2, 3]         

plt.plot(x, y1, color='blue', linestyle='-', marker='o', label='Squares')
plt.plot(x, y2, color='red', linestyle='--', marker='s', label='Linear')
plt.title('Squares vs Linear')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
categories = ['A','B','C']
values = [5, 7, 3]
plt.bar(categories, values)
plt.show()
```

```python
data = [1, 1, 2, 3, 3, 3, 4, 4, 5]
plt.hist(data, bins=10, edgecolor='black')
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [1, 2, 3])
plt.subplot(1, 2, 2)
plt.plot([1, 2, 3], [3, 2, 1])
plt.show()
```
